## Carga el dataset en un DataFrame de Pandas.

## Detecta y filtra outliers usando el método del rango intercuartílico (IQR).

```python
indices_a_eliminar = []

for col in ['Año', 'Cantidad_Migrantes', 'PIB_Origen', 'PIB_Destino', 'IDH_Origen', 'IDH_Destino']:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  limite_inf = Q1 - 1.5 * IQR
  limite_sup = Q3 + 1.5 * IQR
  por_filtrar = df[(df[col] < limite_inf) | (df[col] > limite_sup)]
  if not por_filtrar.empty:
    print(por_filtrar[['Pais_Origen', 'Pais_Destino', col]])
    indices_a_eliminar.extend(por_filtrar.index.tolist())
```

```python
df_filtrado = df.drop(list(set(indices_a_eliminar)))
```

## Reemplaza los valores de la columna "Razon_Migracion" usando mapeo de valores

```python
reemplazos_razon = {
        "Económica": "Trabajo",
        "Conflicto": "Guerra",
        "Educativa": "Estudios"}
df['Razon_Migracion'] = df['Razon_Migracion'].replace(reemplazos_razon)
df.head()
```
